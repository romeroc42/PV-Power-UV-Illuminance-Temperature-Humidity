```python
#Select independent variables from the dependent variable
x = data_mult[["Temperature", "Humidity", "UV", "Illuminance"]]
y = data_mult[["Power"]]
print(x)
print(y)
```

```python
#Create a matrix to represent the x-axis of a graph
x1 = np.arange(0, len(y), 1)
x1
```

```python
#Power vs Time
plt.figure(figsize=(24,10))
plt.title("Power vs Time", fontsize = 16)
plt.scatter(x1, y, s=5, color="red", label="Power")
plt.plot(x1, y, lw=0.8, color="blue", label="Power")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Time', fontsize = 16)
plt.ylabel('Power', fontsize = 16)
plt.legend(fontsize = 16)
plt.show()
```
