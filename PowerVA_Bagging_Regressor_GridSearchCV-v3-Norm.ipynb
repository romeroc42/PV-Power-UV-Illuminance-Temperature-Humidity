#Bagging regressor.

```python
#Tool for plotting
import matplotlib.pyplot as plt
#Wide variety of mathematical operations on arrays
import numpy as np
#SVR Support Vector Regression
from sklearn.svm import SVR
#Bagging regressor is an ensemble meta-estimator
from sklearn.ensemble import BaggingRegressor
#Error metrics and Score metric
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
#Split arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
#Exhaustive search over specified parameter values for an estimator
from sklearn.model_selection import GridSearchCV
#Regression based on k-nearest neighbors
from sklearn.neighbors import KNeighborsRegressor
#Linear Regression and Logistic Regression
from sklearn.linear_model import LinearRegression, LogisticRegression
#Scale input vectors individually to unit norm (vector length)
from sklearn import preprocessing
```

```python
#Select independent variables from the dependent variable
X = data_mult[["Temperature", "Humidity", "UV", "Illuminance"]]
y = data_mult[["Power"]]
```

```python
#Show independent variables
X
```

```python
#Show dependent variable
y
```

```python
#Scale independent variables input vectors individually to unit norm
X = preprocessing.normalize(X)
```

```python
#Convert to dataframe and add coulmns names
X = pd.DataFrame(X)
X.columns = ["Temperature", "Humidity", "UV", "Illuminance"]
```

```python
#Show independent variables
X
```

```python
#Show dependent variable
y
```

```python
#Split dataset for train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
#Verify number of elements in the split
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Set parameters like the estimator, number of estimators and random state
#If None, then the base estimator is a DecisionTreeRegressor
#params = {'base_estimator': SVR(),'n_estimators': 100,'random_state': 123}
params = {'base_estimator': None ,'n_estimators': 100,'random_state': 123}
```

```python
#Apply parameters for Bagging Regressor
bagg_reg = BaggingRegressor(**params)
```

```python
#Fit using training data
bagg_reg.fit(X_train,y_train)
```

```python
#Predict using independent test variables
bagg_predict = bagg_reg.predict(X_test)
print(bagg_predict)
print(bagg_predict.shape)
print(y_test)
print(y_test.shape)
```

```python
#Mean Squared Error
mse = mean_squared_error(y_test, bagg_predict)
print("The mean squared error (MSE) on test set: {:.8f}".format(mse))
```

```python
#Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, bagg_predict))
print("RMSE: %f" % (rmse))
```

```python
#Mean Absolute Error
mae = mean_absolute_error(y_test, bagg_predict)
print("MAE: %f" % (mae))
```

```python
#Mean Absolute Error Percentage
mape = mean_absolute_percentage_error(y_test, bagg_predict)
print("MAPE: %f" % (mape))
```

```python
#Score for training
bagg_score_train = bagg_reg.score(X_train, y_train)
print(bagg_score_train)
```

```python
#Score for test
bagg_score_test = bagg_reg.score(X_test, y_test)
print(bagg_score_test)
```

```python
#Score for test using R2 score function
bagg_r2_score = r2_score(y_test, bagg_predict)
print(bagg_r2_score)
```

```python
#Show Original vs Predicted data
x_ax = range(len(y_test))
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test, lw=0.8, color="blue", label="original")
plt.plot(x_ax, bagg_predict, lw=0.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

```python
#Show Original vs Predicted data
x_ax = range(0,150)
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test[0:150], lw=1.8, color="blue", label="original")
plt.plot(x_ax, bagg_predict[0:150], lw=1.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

#GridSearchCV

```python
#Verify training data
print(X_train.shape[0])
print(X_train.shape[1])
```

```python
#Set hyperparameters
#If the base estimator is None then will use a DecisionTreeRegressor.
%%time

#n_samples = X_train.shape[0]
#n_features = X_train.shape[0]

params = {'base_estimator': [None, LinearRegression(), LogisticRegression(), KNeighborsRegressor(), SVR()],
          'n_estimators': [25,50,100,150],
          #'max_samples': [0.5,1.0, n_samples//2,],
          #'max_features': [0.5,1.0, n_features//2,],
          'max_samples': [0.25,0.5,1.0],
          'max_features': [0.25,0.5,1.0],
          'bootstrap': [True, False],
          'bootstrap_features': [True, False]}

```

```python
#Use GridSearchCV to find the best fit
bagging_regressor_grid = GridSearchCV(BaggingRegressor(random_state=123, n_jobs=-1), param_grid =params, cv=5, n_jobs=-1, verbose=1)
bagging_regressor_grid.fit(X_train, y_train)
```

```python
#Show the best score and model
print('Train R^2 Score : %.3f'%bagging_regressor_grid.best_estimator_.score(X_train, y_train))
print('Test R^2 Score : %.3f'%bagging_regressor_grid.best_estimator_.score(X_test, y_test))
print('Best R^2 Score Through Grid Search : %.3f'%bagging_regressor_grid.best_score_)
print('Best Parameters : ',bagging_regressor_grid.best_params_)
```

```python
#Show the first results from the Cross Validation Table
cross_val_results = pd.DataFrame(bagging_regressor_grid.cv_results_)
print('Number of Various Combinations of Parameters Tried : %d'%len(cross_val_results))

cross_val_results.head()
```
