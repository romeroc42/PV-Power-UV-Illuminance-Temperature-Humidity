# Gradient Boosting Regression

```python
#Tool for plotting
import matplotlib.pyplot as plt
#Wide variety of mathematical operations on arrays
import numpy as np
#Ensemble methods
from sklearn import ensemble
#Permutation importance for feature evaluation
from sklearn.inspection import permutation_importance
#Error metrics and Score metric
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
#Split arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
#Exhaustive search over specified parameter values for an estimator
from sklearn.model_selection import GridSearchCV
```

```python
#Select independent variables from the dependent variable
X = data_mult[["Temperature", "Humidity", "UV", "Illuminance"]]
y = data_mult[["Power"]]
```

```python
#Show independent variables
X
```

```python
#Show dependent variable
y
```

```python
#Split dataset for train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
#Verify number of elements in the split
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Set parameters like the number of estimators and learning rate
params = {'n_estimators': 100,
          'max_depth': 10,
          'min_samples_split': 5,
          'learning_rate': 0.1,
          'loss': 'squared_error'}
```

```python
#Apply parameters for Gradient Boosting Regressor
reg = ensemble.GradientBoostingRegressor(**params)
#Fit using training data
reg.fit(X_train, y_train)
#Predict using independent test variables
reg_predict = reg.predict(X_test)
```

```python
#Mean Squared Error
mse = mean_squared_error(y_test, reg_predict)
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

```python
#Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, reg_predict))
print("RMSE: %f" % (rmse))
```

```python
#Mean Absolute Error
mae = mean_absolute_error(y_test, reg_predict)
print("MAE: %f" % (mae))
```

```python
#Mean Absolute Error Percentage
mape = mean_absolute_percentage_error(y_test, reg_predict)
print("MAPE: %f" % (mape))
```

```python
#Score for training
reg_score_train = reg.score(X_train, y_train)
print(reg_score_train)
```

```python
#Score for test
reg_score_test = reg.score(X_test, y_test)
print(reg_score_test)
```

```python
#Score for test using R2 score function
reg_r2_score = r2_score(y_test, reg_predict)
print(reg_r2_score)
```

```python
#Show Original vs Predicted data
x_ax = range(len(y_test))
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test, lw=0.8, color="blue", label="original")
plt.plot(x_ax, reg_predict, lw=0.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

```python
#Show Original vs Predicted data
x_ax = range(0,150)
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test[0:150], lw=1.8, color="blue", label="original")
plt.plot(x_ax, reg_predict[0:150], lw=1.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

#GridSearchCV

```python
#Verifiy training data
print(X_train.shape[0])
print(X_train.shape[1])
```

```python
#Set hyperparameters
%%time

params = {'n_estimators': [100,150,200,250],
          'max_depth': [4,8,12],
          'min_samples_split': [10,15,20],
          'learning_rate': [0.025,0.05,0.1,0.15],
          'loss': ['squared_error', 'absolute_error', 'huber', 'quantile']}
```

```python
#Use GridSearchCV to find the best fit
grad_boost_regressor_grid = GridSearchCV(ensemble.GradientBoostingRegressor(random_state=123), param_grid =params, cv=5, n_jobs=-1, verbose=1)
grad_boost_regressor_grid.fit(X_train, y_train)

```

```python
#Show the best score and model
print('Train R^2 Score : %.3f'%grad_boost_regressor_grid.best_estimator_.score(X_train, y_train))
print('Test R^2 Score : %.3f'%grad_boost_regressor_grid.best_estimator_.score(X_test, y_test))
print('Best R^2 Score Through Grid Search : %.3f'%grad_boost_regressor_grid.best_score_)
print('Best Parameters : ',grad_boost_regressor_grid.best_params_)
```

```python
#Show the first results from the Cross Validation Table
cross_val_results = pd.DataFrame(grad_boost_regressor_grid.cv_results_)
print('Number of Various Combinations of Parameters Tried : %d'%len(cross_val_results))

cross_val_results.head()
```
