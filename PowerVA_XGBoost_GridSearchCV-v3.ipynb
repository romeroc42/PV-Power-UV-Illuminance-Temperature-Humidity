#XGBoost

```python
#XGBoost
import xgboost as xgb
#Error metrics and Score metric
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
#Wide variety of mathematical operations on arrays
import numpy as np
#Exhaustive search over specified parameter values for an estimator
from sklearn.model_selection import GridSearchCV
#Split arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
#Tool for plotting
import matplotlib.pyplot as plt
```

```python
#Select independent variables from the dependent variable
X = data_mult[["Temperature", "Humidity", "UV", "Illuminance"]]
y = data_mult[["Power"]]
```

```python
#Show independent variables
X
```

```python
#Show dependent variable
y
```

```python
#The input data is stored in a DMatrix object
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
print(data_dmatrix)
```

```python
#Split dataset for train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
#Verify number of elements in the split
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Set parameters like the number of estimators and learning rate
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 10, alpha = 10, n_estimators = 100)
```

```python
#Fit using training data
xg_reg.fit(X_train,y_train)
```

```python
#Predict using independent test variables
preds = xg_reg.predict(X_test)
print(preds)
print(preds.shape)
print(y_test)
print(y_test.shape)
```

```python
#Mean Squared Error
mse = mean_squared_error(y_test, preds)
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```

```python
#Root Mean Squared Error
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
#Mean Absolute Error
mae = mean_absolute_error(y_test, preds)
print("MAE: %f" % (mae))
```

```python
#Mean Absolute Error Percentage
mape = mean_absolute_percentage_error(y_test, preds)
print("MAPE: %f" % (mape))
```

```python
#Score for training
xg_reg_score_train = xg_reg.score(X_train, y_train)
print(xg_reg_score_train)
```

```python
#Score for test
xg_reg_score_test = xg_reg.score(X_test, y_test)
print(xg_reg_score_test)
```

```python
#Score for test using R2 score function
xg_reg_r2_score = r2_score(y_test, preds)
print(xg_reg_r2_score)
```

```python
#Convert specified tree to graphviz instance
xgb.to_graphviz(xg_reg,num_trees=0)
```

```python
#Plot importance based on fitted trees
xgb.plot_importance(xg_reg, max_num_features=15)
#plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```

```python
#Show Original vs Predicted data
x_ax = range(len(y_test))
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test, lw=0.8, color="blue", label="original")
plt.plot(x_ax, preds, lw=0.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

```python
#Show Original vs Predicted data
x_ax = range(0,150)
plt.figure(figsize=(24,10))
plt.title("Original vs Predicted", fontsize = 16)
plt.ylabel("Power", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.plot(x_ax, y_test[0:150], lw=1.8, color="blue", label="original")
plt.plot(x_ax, preds[0:150], lw=1.8, color="red", label="predicted")
plt.legend(fontsize = 16)
plt.show()
```

#GridSearchCV

```python
#Set hyperparameters

%%time

#xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 10, alpha = 10, n_estimators = 100)

params = { 'objective':['reg:squarederror'],
           'colsample_bytree':[0.1,0.3,0.5,0.75],
           'learning_rate':[0.025,0.05,0.1,0.15],
           'max_depth':[4,8,12],
           'alpha':[2.5,5,10],
           'n_estimators':[50,100,150,200,250]}
```

```python
#Use GridSearchCV to find the best fit
xgb_regressor_grid = GridSearchCV(xgb.XGBRegressor(random_state=123), param_grid =params, cv=5, n_jobs=-1, verbose=1)
xgb_regressor_grid.fit(X_train, y_train)
```

```python
#Show the best score and model
print('Train R^2 Score : %.3f'%xgb_regressor_grid.best_estimator_.score(X_train, y_train))
print('Test R^2 Score : %.3f'%xgb_regressor_grid.best_estimator_.score(X_test, y_test))
print('Best R^2 Score Through Grid Search : %.3f'%xgb_regressor_grid.best_score_)
print('Best Parameters : ',xgb_regressor_grid.best_params_)
```

```python
#Show the first results from the Cross Validation Table
cross_val_results = pd.DataFrame(xgb_regressor_grid.cv_results_)
print('Number of Various Combinations of Parameters Tried : %d'%len(cross_val_results))

cross_val_results.head()
```
